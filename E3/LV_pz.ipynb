# The Lotka-Volterra model
This model is a python translation of the LV_pz model proposed in Glover et al. (2011), Sec. 9.3

You will need to make sure the required modules below are installed in your python environment

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

This is the function containing the right hand side of the ODE system

```python
def LV_pz(x,t,p):
    X1,X2 = x
    p1,p2,p3,p4 = p
    dX1dt = X1 * (p1 - p2*X2)
    dX2dt = X2 * (p3*p2*X1 - p4)
    return np.array([dX1dt,dX2dt])
```

## Parameters and initial conditions

```python
# The parameters
p = np.array([0.1,0.4,0.2,0.05])
X1_0 = 0.8
X2_0 = 0.171
x0 = np.array([X1_0,X2_0])
```

## Time axis

```python
# time units are days
Tstart = 0.
Tend = 400.
step = 0.1
t = np.arange(Tstart,Tend,step)
```

## Integration using the scipy ODE solver

```python
# note that the extra argument containing the parameter "args" must be a tuple
y = odeint(LV_pz, x0, t, args=(p,))
```

## Plot Fig. 9.3a

```python
plt.plot(t,y[:,0],'k-',label='Phytoplankton')
plt.plot(t,y[:,1],'k--',label='Zooplankton')
plt.xlabel('Time (days)')
plt.ylabel('Plankton (mmol N/m3)')
plt.legend()
plt.show()
```

By changing P3 =1, there is a clear difference between the plots. The shape changes from a more circular shape to a more triangular shape. By changing p3 to 1, this means that there is a change in the zooplankton's assimilation of the phytoplankton. There is now a 1-1 assimilation, meaning the zooplankton take up the phytoplankton completely. This means that all the mass in the system, is contained within the system. We see this by the more linear relationship between phytoplankton and zooplankton observed in the figure however, the relationship is unstable in this situation. The zooplankton assimilate all the phytoplankton, causing their population to grow rapidly until the phytoplankton are depleted. The system then resets with the phytoplankton growing, and then the zooplankton growing rapidly by the assimilation of this phytoplankton but then the system gets depleted again.

Here's the full code you can use in Jupyter Lab to generate the Lotka-Volterra phase-space diagram using a scatter plot:
import numpy as np
import matplotlib.pyplot as plt

# Lotka-Volterra equations
def lotka_volterra(prey, predator, a, b, c, d):
    dprey_dt = a * prey - b * prey * predator
    dpredator_dt = c * prey * predator - d * predator
    return dprey_dt, dpredator_dt

# Parameters
a = 1.1  # Prey (phytoplankton) birth rate
b = 0.4  # Predation rate
c = 0.1  # Predator (zooplankton) reproduction rate
d = 0.4  # Predator death rate

# Time settings
dt = 0.01
time = np.arange(0, 50, dt)

# Initial populations
prey = np.zeros_like(time)
predator = np.zeros_like(time)
prey[0] = 10
predator[0] = 5

# Simulate dynamics
for t in range(1, len(time)):
    dprey_dt, dpredator_dt = lotka_volterra(prey[t-1], predator[t-1], a, b, c, d)
    prey[t] = prey[t-1] + dprey_dt * dt
    predator[t] = predator[t-1] + dpredator_dt * dt

# Phase-space scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(prey, predator, s=1, color='blue')
plt.title('Lotka-Volterra Phase-Space Diagram')
plt.xlabel('Phytoplankton (Prey)')
plt.ylabel('Zooplankton (Predator)')
plt.grid(True)
plt.show()

What this does:

Simulates the predator-prey dynamics over time.
Plots a scatter plot of phytoplankton vs. zooplankton values at each time step.
Shows the cyclical nature of the interaction and the attractor behavior.
