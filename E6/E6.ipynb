```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#A numerical solution of the decay equation 
import pylab as pl

t0 = 0. # time is in days
tn = 20.
dt = 10.
kappa = 0.05 # day−1
C0 = 0.5 # mg chl /m3

def F(k,C) : # define the function for the right hand side of the equation
    return - k*C
    
# compute the total number of steps
Ntot = ( pl . floor ((tn - t0)/dt) + 1) . astype( int )

# initialization
C = pl . zeros (Ntot)
t = C. copy() # this array stores the time and has the same size as C
t [0] = t0
C[0] = C0

# start the loop
for n in range(Ntot - 1):
    C[n+1] = C[n] + F(kappa ,C[n]) * dt
    t [n+1] = t [n] + dt
```

```python
plt.figure()
plt.plot(t, C, marker='x', linestyle='-', color='pink')
plt.xlabel('Time (days)')
plt.ylabel('Concentration (mg chl/m3)')
plt.title('Euler-forward decay ODE')
plt.show()
```

## Euler-backward solution: ##

```python
#Loading the parameters: 
t0 = 0. # time is in days
tn = 20.
dt = 10.
kappa = 0.05 # day−1
C0 = 0.5 # mg chl /m3
    
# compute the total number of steps
Ntot = ( pl . floor ((tn - t0)/dt) + 1) . astype( int )

# initialization
C = pl . zeros (Ntot)
t = C. copy() # this array stores the time and has the same size as C
t [0] = t0
C[0] = C0

# start the loop
for n in range(Ntot - 1):
    C[n+1] = C[n] / ( 1 + kappa * dt)
    t [n+1] = t [n] + dt
```

```python
plt.figure()
plt.plot(t, C, marker='x', linestyle='-', color='pink')
plt.xlabel('Time (days)')
plt.ylabel('Concentration (mg chl/m3)')
plt.title('Euler-backward decay ODE')
plt.show()
```

```python
#Loading the parameters: 
t0 = 0. # time is in days
tn = 20.
dt = 10.
kappa = 0.05 # day−1
C0 = 0.5 # mg chl /m3
    
# compute the total number of steps
Ntot = ( pl . floor ((tn - t0)/dt) + 1) . astype( int )

# initialization
C = pl . zeros (Ntot)
t = C. copy() # this array stores the time and has the same size as C
t [0] = t0
C[0] = C0

# start the loop
for n in range(Ntot - 1):
    C[n+1] = C[n] / ( 1 + kappa * dt)
    t [n+1] = t [n] + dt
    
#### dt = 5
t0 = 0.
tn = 20.
dt1 = 5.
kappa = 0.05
C0 = 0.5

Ntot1 = int(np.floor((tn - t0) / dt1) + 1)

C1 = np.zeros(Ntot1)
t1 = np.zeros(Ntot1)
C1[0] = C0
t1[0] = t0

for n in range(Ntot1 - 1):
    C1[n+1] = C1[n] / (1 + kappa * dt1)
    t1[n+1] = t1[n] + dt1

#### dt = 15
dt2 = 15.
Ntot2 = int(np.floor((tn - t0) / dt2) + 1)

C2 = np.zeros(Ntot2)
t2 = np.zeros(Ntot2)
C2[0] = C0
t2[0] = t0

for n in range(Ntot2 - 1):
    C2[n+1] = C2[n] / (1 + kappa * dt2)
    t2[n+1] = t2[n] + dt2
```

```python
#Plotting: 
plt.figure()
plt.plot(t, C, marker='x', linestyle='-', color='pink', label='dt = 10')  
plt.plot(t1, C1, marker='x', linestyle='-', color='blue', label='dt = 5')
plt.plot(t2, C2, marker='x', linestyle='-', color='green', label='dt = 15')
plt.xlabel('Time (days)')
plt.ylabel('Concentration (mg chl/m³)')
plt.title('Euler-Backward Decay ODE for Different dt Values')
plt.legend()
plt.tight_layout()
plt.show()
```

By changing the dt value, there are changes in the Euler-backward graphs. For a much larger dt, there is a more linear relationship, with the graph ending sooner than the other plots. For the dt=5 line, there is a less steep graph with a less steep gradient reported for a smaller dt. Due to the small timestep and dt value, the line for dt=15 ends at t=15, as it is not possible for the graph to proceed further, however a higher gradient is present compared to others. As the dt decreases, there seems to be more steps in the line as well as a less linear relationship compared to the lines with higher dt values.
