```python
#Importing the packages:
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
#Plotting the figure:

plt.figure()
plt.plot(t, P_function_t, label = "Phytoplankton", color = 'blue')
plt.plot(t, N_function_t,label = "Nutrients", color = 'red')
plt.xlabel('Time (days)')
plt.ylabel('Concentration (mmol N/m³)')
plt.title('Phytoplankton growth over time')
plt.tight_layout()
plt.legend()
plt.show()
```

The curves are identical in shape as they follow the same pattern. The phytoplankton are dependant on the nutrients for their growth and there is a relationship between them. However, it is not the exact same curves but they are inverses of each other. As phytoplankton grow, they consume the nutrients. Therefore, when the phytoplankton are at a low concentration, the nurtients are at a high concentration. As the population grows exponentially, there is an exponential decay of the nutrients as they are used at the same rate of the growth. As nutrients become depleted and the curve plateaus. There is then a plateau of the phytoplankton growth curve as their growth is now limited by the lack of nutrients. 

*Question 2: What happens when you change the value of the uptake parameter?*

```python
#Adjusting the uptake rate:

k_prime_adjusted = 0.5 #d-1
P0 = 0.1 #mmol N m-3
N0 = 5 #mmol N m-3


t = np.linspace(0, 20, 400)

#Equations + plotting: 
exponent = np.exp(-k_prime_adjusted * (P0 + N0) * t)
P_function_t = P0 * ((P0 + N0) / (P0 + N0 * exponent))
N_function_t = (N0 + P0) - P_function_t

plt.figure()
plt.plot(t, P_function_t, label = "Phytoplankton", color = 'blue')
plt.plot(t, N_function_t,label = "Nutrients", color = 'red')
plt.xlabel('Time (days)')
plt.ylabel('Concentration (mmol N/m³)')
plt.title('Phytoplankton growth over time')
plt.tight_layout()
plt.legend()
plt.show()
```

By increasing the uptake rate, there is a large increase in the steepness of the graph, for phytoplankton growth. This indicates that more food is taken up per day by the phytoplankton when the uptake rate is increased. However, there is a plateau much sooner compared to the other graph which means that the population growth is limited sooner and for longer. Therefore, by increasing the uptake rate, this causes faster growth/consumption overall but the population growth is inhibited sooner. Similarly, the nutrients graph follows the same pattern. It starts at a high concentration and then rapidly decreases  and plateaus sooner. This means that the population is nutrient limited much sooner.
