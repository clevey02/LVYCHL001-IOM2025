Importing the required packages:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('flynn94.csv')
```

```python
print(df.head())
```

```python
#Plotting:
plt.figure()
plt.plot(df['time'], df['cells'], marker='x') 
plt.xlabel('Time')
plt.ylabel('Cells')
plt.title('Population growth over time')
plt.tight_layout()
plt.show()
```

*Question 1: Is the growth exponential?*

The growth appears to be mostly exponential, especially during the early to middle phase of the graph. However, after approximately time t = 18, the growth begins to plateau. Initially, from time 0 to 5, there is no immediate exponential increase—the curve remains relatively flat just above zero. This suggests that in the beginning, certain limiting factors restrict the population growth. Once some initial conditions are met, exponential growth seems to occur. However, after the population reaches a threshold, other factors appear to inhibit further growth, resulting in the plateau.

*Question 2: Can you separate the growth curve into some phases?*

Yes, there are three visible phases to the above curve. Initially, phase 1, from time 0-5, there appears to be minimal growth in the population. This phase can be described as the intial population growth, where the growth overall is slow. Phase 2, from time 5-±18, is where there is exponential growth of the population. This is due to the rapid growth of the population, after some conditions have been met after Phase 1, the population is able to growth exponentially. After time =18, Phase 3 occurs. In this phase, there seems to be other factors that limit the population growth above some kind threshold value of the population (Cells = 2.00). This could mean that above a certain population size, other factors limit the population growth. 

*Question 3: How different is the growth expressed in cellular carbon from the one in
cell number and chlorophyll?*

```python
#Plotting Carbon: 
plt.figure()
plt.plot(df['time'], df['cellC'], marker='x') 
plt.xlabel('Time')
plt.ylabel('Cellular Carbon')
plt.title('Population growth over time, expressed with cellular Carbon')
plt.tight_layout()
plt.show()
```

```python
#Plotting Chlorophyll: 
plt.figure()
plt.plot(df['time'], df['chl-a'], marker='x') 
plt.xlabel('Time')
plt.ylabel('Chlorophyll-a')
plt.title('Population growth over time, expressed with chlorophyll-a')
plt.tight_layout()
plt.show()
```

The plot expressed with Cellular Carbon, shows a similar shape to the one with cells. Both graphs display a slow growth rate intitially, followed by the exponential growth of the population and then the plateau of the population after some threshold value fo the population. However, the gradient is less steep with regards to the Carbon graph and there are a few data points missing from this graph. The intial flat region and plateau are less pronounced for the carbon graph than the cell graph as well, which could be related to the gradient of the graph. This could be because although the population no longer continues growing, the phytoplankton continue producing carbon.  

Comparing the plots of the cellular carbon and chlorophyll-a, the plots seem to show some similarities and some differences. The similarities include a slow growth rate initially, followed by an exponetial growth. The chlorophyll-a graph however, seems to have a steeper gradient of exponential growth compared to the carbon plot, as well as a longer initial lag time. Another difference is a sharp drop off (decrease) in chlorophyll-a after time = ±18. Whereas at this time for carbon, there is still an increase in carbon, and no plateau is visible. The sharp drop in chlorophyll-a could indicate that above the population threshold value, some cells/chloroplasts begin to die, whereas carbon is still being produced. 

*Question 4: Is the ammonium consumption approximated by an exponential curve?*

```python
#Plotting Ammonium: 
plt.figure()
plt.plot(df['time'], df['ammonium'], marker='x') 
plt.xlabel('Time')
plt.ylabel('Ammonium Concentration')
plt.title('Ammonium consumption over time')
plt.tight_layout()
plt.show()
```

This curve shows a negative exponential trend, showing exponential decay rather than growth. Initially, there is a gradual decline in ammonium concentration, followed by a rapid decrease, which likely corresponds to a period of rapid population growth—during which ammonium is consumed more quickly. As the population grows and consumes more ammonium, the concentration drops sharply. However, after time = ±12, the ammonium concentration falls to 0 and is plateaud for the rest of the time period. This indicates that the nutrient is depleted and the population can no longer assimilate it. This could mean that this is the limiting factor of the population growth.
